```python
import pandas as pd
df = pd.read_csv("data/patients.csv")
df.head()
```

```python
import numpy as np

tipos_sangre = ["A+", "A-", "B+", "B-", "AB+", "AB-", "O+", "O-"]
probabilidades_sangre = [0.34, 0.06, 0.10, 0.02, 0.04, 0.01, 0.39, 0.04]

df['Tipo de Sangre'] = np.random.choice(tipos_sangre, size=len(df), p=probabilidades_sangre)

posibles_alergias = [
    'Ninguna',
    'Penicilina',
    'Polen',
    'Ácaros del polvo',
    'Mariscos',
    'Frutos secos',
    'Látex',
    'Picaduras de insectos',
    'Gluten',
    'Lactosa',
    'Medicamentos antiinflamatorios',
    'Animales domésticos',
    'Moho'
]

def generar_alergias():
    num_alergias = np.random.choice([0, 1, 2, 3], p=[0.5, 0.3, 0.15, 0.05])
    if num_alergias == 0:
        return 'Ninguna'
    else:
        alergias_seleccionadas = np.random.choice(posibles_alergias[1:], size=num_alergias, replace=False)
        return ', '.join(alergias_seleccionadas)

df['Alergias'] = df.apply(lambda x: generar_alergias(), axis=1)

df[['Nombre', 'Tipo de Sangre', 'Alergias']].head()
```

```python
df.to_csv("data/updated_patients_records_1000.csv", index=False)
```

```python
# encrypt_curp.py
import qrcode

def encrypt_curp(curp, shift=3):
    encrypted = ''
    for char in curp:
        if char.isupper():
            encrypted += chr((ord(char) + shift - 65) % 26 + 65)
        elif char.islower():
            encrypted += chr((ord(char) + shift - 97) % 26 + 97)
        elif char.isdigit():
            encrypted += chr((ord(char) + shift - 48) % 10 + 48)
        else:
            encrypted += char  # No modifica caracteres especiales
    return encrypted

def encrypt_curp_and_generate_qr(curp):
    # Encriptar el CURP usando el cifrado César
    encrypted_curp = encrypt_curp(curp)

    # Generar el código QR con el CURP encriptado
    qr = qrcode.QRCode(version=1, box_size=10, border=4)
    qr.add_data(encrypted_curp)
    qr.make(fit=True)

    img = qr.make_image(fill='black', back_color='white')

    # Guardar la imagen del código QR
    img.save(f"qr_{curp}.png")

    print(f"Código QR generado y guardado como qr_{curp}.png")

if __name__ == "__main__":
    # Uso de ejemplo
    curp_to_encrypt = "PA040812MNL7281642"  # Reemplaza con el CURP que desees encriptar
    encrypt_curp_and_generate_qr(curp_to_encrypt)
```
